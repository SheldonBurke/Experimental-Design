**Question 4. Choose any three ethnicities and compare data for the states in Part 1 using scatter plots**

```python
sns.scatterplot(data=county_data, x="Ethnicities.White Alone", y="Population.2010 Population",
hue="State",palette="rocket", legend="full" ).set(title="White population in States by Counties",xlabel='White population as % in county', ylabel='Population in county')
plt.ticklabel_format(style='plain', axis='y') #takes the y axis out of scientic notation


```

```python
sns.scatterplot(data=county_data, x="Ethnicities.Black Alone", y="Population.2010 Population",
hue="State",palette="copper", legend="full" ).set(title="Black population in States by Counties",xlabel='Black population as % in county', ylabel='Population in county')
plt.ticklabel_format(style='plain', axis='y') #takes the y axis out of scientic notation
```

```python
sns.scatterplot(data=county_data, x="Ethnicities.American Indian and Alaska Native Alone", y="Population.2010 Population",
hue="State",palette="gist_earth", legend="full" ).set(title="Native population in States by Counties",xlabel=' Native population as % in county', ylabel='Population in county')
plt.ticklabel_format(style='plain', axis='y') #takes the y axis out of scientic notation
```

Question 5. Compare the data in the two columns (Population.2020 Population and Population.2010
Population) for each of the two states in Part 1using histogram.

```python
sns.histplot(data=county_data, x='Population.2010 Population', hue='State', multiple='dodge', shrink=0.8, element='step', bins=10, palette="bone_r").set(title="County Population in 2010", xlabel=' Population in County', ylabel='Number of counties in range')
#used step to avoid overlapping
plt.ticklabel_format(style='plain', axis='x') #takes the y axis out of scientic notation
```

```python
sns.histplot(data=county_data, x='Population.2020 Population', hue='State', multiple='dodge', shrink=0.8, element='step', bins=10, palette="cividis").set(title="County Population in 2020", xlabel=' Population in County', ylabel='Number of counties in range')
#used step to avoid overlapping
plt.ticklabel_format(style='plain', axis='x') #takes the y axis out of scientic notation
```
